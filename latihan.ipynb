```python
import pandas as pd
#baca data
df=pd.read_csv('day.csv')

#Tampilin data
df.head()
```

```python
#cek info data yang bolong
df.info()
```

```python
import matplotlib.pyplot as plt
#Ubah tanggal jadi format kalender
df['dteday']= pd.to_datetime(df['dteday'])

#Gambar grafik
plt.figure(figsize=(12,5))
plt.plot(df['dteday'], df['cnt'], color='blue')
plt.title("Grafik Jumlah Penyewa sepeda (2011-2012)")
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Penyewa')
plt.show()
```
